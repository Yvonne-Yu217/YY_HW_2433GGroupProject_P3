## 4. Data Compatibility Check

## 3. Load State Mapping from Codebook

To join the two datasets, we need to convert CDC's `_STATE` column (FIPS codes) to state names that match KFF's `Location` format.

```python
# Load state FIPS code mapping from codebook
state_mapping_path = Path('2433_p3_data/healthcare.gov/exports/codebook_tables/state_fips_codes.csv')

if state_mapping_path.exists():
    state_mapping_df = pd.read_csv(state_mapping_path)
    # Skip the first row which contains the header description
    state_mapping_df = state_mapping_df[state_mapping_df['value'] != 'Label:'].copy()
    state_mapping_df['value'] = pd.to_numeric(state_mapping_df['value'], errors='coerce')
    state_mapping_df = state_mapping_df.dropna(subset=['value'])
    state_mapping_df['value'] = state_mapping_df['value'].astype(int)
    
    # Create a dictionary for mapping FIPS code to state name
    state_fips_to_name = dict(zip(state_mapping_df['value'], state_mapping_df['label']))
    
    print(f"âœ“ Loaded state FIPS mapping")
    print(f"  Total states/territories: {len(state_fips_to_name)}")
    print(f"\nSample mappings:")
    for fips, name in list(state_fips_to_name.items())[:5]:
        print(f"  {fips:2d} â†’ {name}")
else:
    print(f"âœ— File not found: {state_mapping_path}")
    state_fips_to_name = {}
```

```python
# Apply state name mapping to CDC data
if 'cdc_df' in globals() and state_fips_to_name:
    # Create new 'Location' column in CDC data by mapping _STATE to state names
    cdc_df['Location'] = cdc_df['_STATE'].map(state_fips_to_name)
    
    print(f"âœ“ Added 'Location' column to CDC data")
    print(f"  Mapped {cdc_df['Location'].notna().sum():,} out of {len(cdc_df):,} rows")
    print(f"  Unmapped FIPS codes: {cdc_df['Location'].isna().sum()}")
    
    # Show sample of the mapping
    print(f"\nSample CDC data with Location:")
    display(cdc_df[['_STATE', 'Location', 'IYEAR', 'Age_Group']].head(10))
    
    # Check unique locations in both datasets
    if 'kff_df' in globals():
        kff_locations = set(kff_df['Location'].unique())
        cdc_locations = set(cdc_df['Location'].dropna().unique())
        common_locations = kff_locations & cdc_locations
        
        print(f"\nðŸ“Š Location overlap analysis:")
        print(f"  KFF unique locations: {len(kff_locations)}")
        print(f"  CDC unique locations: {len(cdc_locations)}")
        print(f"  Common locations: {len(common_locations)}")
        
        # Show locations that don't overlap
        kff_only = kff_locations - cdc_locations
        cdc_only = cdc_locations - kff_locations
        if kff_only:
            print(f"\n  KFF-only locations: {sorted(kff_only)}")
        if cdc_only:
            print(f"  CDC-only locations: {sorted(cdc_only)}")
else:
    print("âš  CDC data or state mapping not available")
```

```python
# Check common dimensions for potential joins
if 'kff_df' in globals() and 'cdc_df' in globals():
    print("=" * 100)
    print("Data Compatibility Analysis")
    print("=" * 100)
    
    # Check for common columns
    kff_cols = set(kff_df.columns)
    brfss_cols = set(cdc_df.columns)
    common_cols = kff_cols & brfss_cols
    
    print(f"\nCommon columns: {common_cols if common_cols else 'None'}")
    
    # Check for potential join keys
    potential_keys = ['Year', 'year', 'State', 'state', 'Location', 'location', '_STATE', 'StateCode']
    print(f"\nPotential join keys:")
    print(f"  KFF columns: {[col for col in kff_df.columns if any(k.lower() in col.lower() for k in potential_keys)]}")
    print(f"  CDC columns: {[col for col in cdc_df.columns if any(k.lower() in col.lower() for k in potential_keys)]}")
    
    # Check year ranges
    kff_year_col = [col for col in kff_df.columns if 'year' in col.lower()]
    brfss_year_col = [col for col in cdc_df.columns if 'year' in col.lower()]
    
    if kff_year_col and brfss_year_col:
        kff_years = sorted(kff_df[kff_year_col[0]].dropna().unique())
        brfss_years = sorted(cdc_df[brfss_year_col[0]].dropna().unique())
        overlapping_years = set(kff_years) & set(brfss_years)
        
        print(f"\nYear ranges:")
        print(f"  KFF: {kff_years}")
        print(f"  CDC: {brfss_years}")
        print(f"  Overlapping years: {sorted(overlapping_years) if overlapping_years else 'None'}")
```

# Part 2: Time-Lagged Prediction Model with Metal Tier

## Objective
Build a predictive model where:
- **X (Features)**: CDC health risk data (Year T) + Metal Tier (Bronze/Silver/Gold)
- **Y (Target)**: KFF premium prices (Year T+2)

### Business Logic
Insurance companies use historical health data to forecast future risk and set premiums with a ~2-year lag. Different metal tiers have different pricing structures.

### Key Data Transformation
Convert KFF data from **Wide Format** (one row per state) to **Long Format** (one row per state Ã— metal tier).

## 7. Transform KFF Data: Wide â†’ Long Format (with Metal Tiers)

```python
# Convert KFF data from Wide to Long format
# Current: Location | Bronze_Price | Silver_Price | Gold_Price | Year
# Target: Location | Metal_Tier | Premium | Year

# First, let's see the current column names
print("Original KFF columns:")
print(kff_df.columns.tolist())
print("\nOriginal shape:", kff_df.shape)
print("\nSample data:")
kff_df.head(3)
```

```python
# Melt the KFF data to long format
kff_long = pd.melt(
    kff_df,
    id_vars=['Location', 'Year'],
    value_vars=['Average Lowest-Cost Bronze Premium', 
                'Average Lowest-Cost Silver Premium', 
                'Average Benchmark Premium', 
                'Average Lowest-Cost Gold Premium'],
    var_name='Metal_Tier_Raw',
    value_name='Premium_Y'
)

print("After melt - Shape:", kff_long.shape)
print("\nSample melted data:")
kff_long.head(10)
```

```python
# Clean Metal_Tier column - extract tier name from raw column name
def extract_metal_tier(tier_raw):
    if 'Bronze' in tier_raw:
        return 'Bronze'
    elif 'Silver' in tier_raw:
        return 'Silver'
    elif 'Gold' in tier_raw:
        return 'Gold'
    elif 'Benchmark' in tier_raw:
        return 'Silver'  # Benchmark is typically Silver tier
    else:
        return 'Unknown'

kff_long['Metal_Tier'] = kff_long['Metal_Tier_Raw'].apply(extract_metal_tier)

# Drop the raw column
kff_long = kff_long.drop('Metal_Tier_Raw', axis=1)

print("After cleaning Metal_Tier:")
print("\nMetal Tier distribution:")
print(kff_long['Metal_Tier'].value_counts())
print("\nSample cleaned data:")
kff_long.head(10)
```

```python
# Convert Premium_Y from string ($XXX) to numeric
kff_long['Premium_Y'] = kff_long['Premium_Y'].str.replace('$', '').str.replace(',', '').astype(float)

print("After converting Premium_Y to numeric:")
print("\nPremium_Y statistics:")
print(kff_long['Premium_Y'].describe())
print("\nCheck for missing values:")
print(kff_long.isnull().sum())
print("\nFinal Long Format KFF Data:")
print(kff_long.head(15))  # Show 15 rows to see multiple tiers
```

## 8. Rebuild Time-Lagged Dataset with Metal Tier as Feature

```python
# First check what columns exist in cdc_df
print("CDC DataFrame columns:")
print(cdc_df.columns.tolist())
print("\nCDC DataFrame shape:", cdc_df.shape)
cdc_df.head(3)
```

```python
# Prepare CDC features (use IYEAR as the year column)
cdc_features = cdc_df[['IYEAR', 'Location', 'Age_Group', 
                        'mean_BMI_w', 'ever_smoker_prev_w', 'current_smoker_prev_w',
                        'diabetes_prev_w', 'heart_attack_prev_w', 'chd_prev_w', 
                        'stroke_prev_w', 'asthma_prev_w', 'copd_prev_w', 
                        'arthritis_prev_w', 'has_children_prev_w', 
                        'mean_num_adults_w', 'mean_household_size_w']].copy()

# Rename IYEAR to CDC_Year for clarity
cdc_features.rename(columns={'IYEAR': 'CDC_Year'}, inplace=True)

# Create time-lagged version: CDC Year T â†’ KFF Year T+2
cdc_lagged = cdc_features.copy()
cdc_lagged['KFF_Year'] = cdc_lagged['CDC_Year'] + 2

print("CDC lagged data prepared:")
print(f"Shape: {cdc_lagged.shape}")
print(f"CDC Years: {sorted(cdc_lagged['CDC_Year'].unique())}")
print(f"Target KFF Years: {sorted(cdc_lagged['KFF_Year'].unique())}")
print("\nSample:")
cdc_lagged.head()
```

```python
# Now merge with kff_long to include Metal_Tier as a feature
# Join on Location, KFF_Year (from CDC) = Year (from KFF)
modeling_df_v2 = pd.merge(
    cdc_lagged,
    kff_long,
    left_on=['Location', 'KFF_Year'],
    right_on=['Location', 'Year'],
    how='inner'
)

# Drop the redundant Year column from KFF
modeling_df_v2 = modeling_df_v2.drop('Year', axis=1)

print("New modeling dataset with Metal_Tier:")
print(f"Shape: {modeling_df_v2.shape}")
print(f"\nColumns: {modeling_df_v2.columns.tolist()}")
print(f"\nYear distribution:")
print(f"CDC Years: {sorted(modeling_df_v2['CDC_Year'].unique())}")
print(f"KFF Years: {sorted(modeling_df_v2['KFF_Year'].unique())}")
print(f"\nMetal Tier distribution:")
print(modeling_df_v2['Metal_Tier'].value_counts())
print(f"\nMissing values:")
print(modeling_df_v2.isnull().sum())
print("\nSample rows:")
modeling_df_v2.head(10)
```

## 9. Train/Val/Test Split with Metal Tier

```python
# Split by KFF year (same strategy as before)
# Train: 2022-2024 premiums (from CDC 2020-2022)
# Val: 2025 premiums (from CDC 2023)
# Test: 2026 premiums (from CDC 2024) - critical evaluation set

train_df_v2 = modeling_df_v2[modeling_df_v2['KFF_Year'].isin([2022, 2023, 2024])].copy()
val_df_v2 = modeling_df_v2[modeling_df_v2['KFF_Year'] == 2025].copy()
test_df_v2 = modeling_df_v2[modeling_df_v2['KFF_Year'] == 2026].copy()

print("Dataset split:")
print(f"Train (2022-2024): {train_df_v2.shape[0]} rows")
print(f"Val (2025): {val_df_v2.shape[0]} rows")
print(f"Test (2026): {test_df_v2.shape[0]} rows")
print(f"Total: {train_df_v2.shape[0] + val_df_v2.shape[0] + test_df_v2.shape[0]} rows")

print("\nMetal Tier distribution per set:")
print("\nTrain:")
print(train_df_v2['Metal_Tier'].value_counts())
print("\nVal:")
print(val_df_v2['Metal_Tier'].value_counts())
print("\nTest:")
print(test_df_v2['Metal_Tier'].value_counts())
```

## 10. Feature Engineering V2: Include Metal_Tier

```python
# Define features including Metal_Tier
# 13 numerical health features + 2 categorical features (Age_Group, Metal_Tier)

numerical_features = [
    'mean_BMI_w', 'ever_smoker_prev_w', 'current_smoker_prev_w',
    'diabetes_prev_w', 'heart_attack_prev_w', 'chd_prev_w', 
    'stroke_prev_w', 'asthma_prev_w', 'copd_prev_w', 
    'arthritis_prev_w', 'has_children_prev_w', 
    'mean_num_adults_w', 'mean_household_size_w'
]

categorical_features = ['Age_Group', 'Metal_Tier']

feature_cols = numerical_features + categorical_features
target_col = 'Premium_Y'

print(f"Total features: {len(feature_cols)}")
print(f"  - Numerical: {len(numerical_features)}")
print(f"  - Categorical: {len(categorical_features)}")
print(f"\nFeature list: {feature_cols}")

# Check missing values in features
print("\nMissing value percentage in training set:")
missing_pct = (train_df_v2[feature_cols].isnull().sum() / len(train_df_v2) * 100).sort_values(ascending=False)
print(missing_pct[missing_pct > 0])
```

## 11. Preprocessing V2: Handle Missing Values + Encode Categoricals

```python
from sklearn.preprocessing import LabelEncoder

# Separate X and y for train/val/test
X_train_v2 = train_df_v2[feature_cols].copy()
y_train_v2 = train_df_v2[target_col].copy()

X_val_v2 = val_df_v2[feature_cols].copy()
y_val_v2 = val_df_v2[target_col].copy()

X_test_v2 = test_df_v2[feature_cols].copy()
y_test_v2 = test_df_v2[target_col].copy()

print("Before encoding:")
print(f"X_train shape: {X_train_v2.shape}")
print(f"X_val shape: {X_val_v2.shape}")
print(f"X_test shape: {X_test_v2.shape}")

# Encode categorical features (Age_Group and Metal_Tier)
# Age_Group: already numeric (1.0, 2.0, 3.0, etc.)
# Metal_Tier: needs encoding (Bronze, Silver, Gold)

le_metal = LabelEncoder()
le_metal.fit(['Bronze', 'Silver', 'Gold'])  # Fixed order for consistency

X_train_v2['Metal_Tier_Encoded'] = le_metal.transform(X_train_v2['Metal_Tier'])
X_val_v2['Metal_Tier_Encoded'] = le_metal.transform(X_val_v2['Metal_Tier'])
X_test_v2['Metal_Tier_Encoded'] = le_metal.transform(X_test_v2['Metal_Tier'])

# Drop original Metal_Tier column
X_train_v2 = X_train_v2.drop('Metal_Tier', axis=1)
X_val_v2 = X_val_v2.drop('Metal_Tier', axis=1)
X_test_v2 = X_test_v2.drop('Metal_Tier', axis=1)

print("\nMetal Tier encoding:")
print(f"Bronze â†’ {le_metal.transform(['Bronze'])[0]}")
print(f"Silver â†’ {le_metal.transform(['Silver'])[0]}")
print(f"Gold â†’ {le_metal.transform(['Gold'])[0]}")

print("\nAfter encoding:")
print(f"X_train shape: {X_train_v2.shape}")
print(f"Column names: {X_train_v2.columns.tolist()}")
```

```python
# Impute missing values and standardize
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

# Imputer for missing values (use median strategy)
imputer = SimpleImputer(strategy='median')
imputer.fit(X_train_v2)

X_train_imputed = imputer.transform(X_train_v2)
X_val_imputed = imputer.transform(X_val_v2)
X_test_imputed = imputer.transform(X_test_v2)

# Standardize features (mean=0, std=1)
scaler = StandardScaler()
scaler.fit(X_train_imputed)

X_train_scaled = scaler.transform(X_train_imputed)
X_val_scaled = scaler.transform(X_val_imputed)
X_test_scaled = scaler.transform(X_test_imputed)

print("Preprocessing complete:")
print(f"X_train_scaled shape: {X_train_scaled.shape}")
print(f"X_val_scaled shape: {X_val_scaled.shape}")
print(f"X_test_scaled shape: {X_test_scaled.shape}")
print(f"\ny_train shape: {y_train_v2.shape}")
print(f"y_val shape: {y_val_v2.shape}")
print(f"y_test shape: {y_test_v2.shape}")
```

## 12. XGBoost Model V2: With Metal_Tier Feature and Regularization

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import time

# Train XGBoost model with regularization to prevent overfitting
xgb_model_v2 = XGBRegressor(
    n_estimators=200,          # More trees but with slower learning
    max_depth=5,               # Shallower trees to reduce overfitting
    learning_rate=0.05,        # Slower learning rate
    subsample=0.8,             # Use 80% of data for each tree
    colsample_bytree=0.8,      # Use 80% of features for each tree
    reg_alpha=0.1,             # L1 regularization
    reg_lambda=1.0,            # L2 regularization
    random_state=42,
    n_jobs=-1,
    early_stopping_rounds=20,  # Stop if no improvement for 20 rounds
    eval_metric='rmse'
)

print("Training XGBoost V2 (with Metal_Tier and regularization)...")
print("Model configuration:")
print(f"  n_estimators: 200")
print(f"  max_depth: 5 (shallower to prevent overfitting)")
print(f"  learning_rate: 0.05 (slower learning)")
print(f"  subsample: 0.8")
print(f"  reg_alpha (L1): 0.1")
print(f"  reg_lambda (L2): 1.0")

start_time = time.time()
xgb_model_v2.fit(
    X_train_scaled, 
    y_train_v2,
    eval_set=[(X_val_scaled, y_val_v2)],
    verbose=False
)
train_time = time.time() - start_time
print(f"\nTraining complete in {train_time:.2f} seconds")

# Make predictions
y_train_pred_v2 = xgb_model_v2.predict(X_train_scaled)
y_val_pred_v2 = xgb_model_v2.predict(X_val_scaled)
y_test_pred_v2 = xgb_model_v2.predict(X_test_scaled)

print("Predictions generated for all sets.")
```

```python
# Calculate metrics
def calculate_metrics(y_true, y_pred, set_name):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    
    print(f"\n{set_name} Set Performance:")
    print(f"  MAE: ${mae:.2f}")
    print(f"  RMSE: ${rmse:.2f}")
    print(f"  RÂ²: {r2:.4f}")
    print(f"  MAPE: {mape:.2f}%")
    
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}

train_metrics_v2 = calculate_metrics(y_train_v2, y_train_pred_v2, "Train")
val_metrics_v2 = calculate_metrics(y_val_v2, y_val_pred_v2, "Validation")
test_metrics_v2 = calculate_metrics(y_test_v2, y_test_pred_v2, "Test")

# Display model performance summary
print("\n" + "="*60)
print("XGBoost Model Performance (with Metal_Tier)")
print("="*60)
print("\nModel Results:")
print(f"  Train RÂ²: {train_metrics_v2['R2']:.2f}, Test RÂ²: {test_metrics_v2['R2']:.2f}, Test MAPE: {test_metrics_v2['MAPE']:.2f}%")
```
