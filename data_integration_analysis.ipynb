## 4. Data Compatibility Check

## 3. Load State Mapping from Codebook

To join the two datasets, we need to convert CDC's `_STATE` column (FIPS codes) to state names that match KFF's `Location` format.

```python
# Load state FIPS code mapping from codebook
state_mapping_path = Path('2433_p3_data/healthcare.gov/exports/codebook_tables/state_fips_codes.csv')

if state_mapping_path.exists():
    state_mapping_df = pd.read_csv(state_mapping_path)
    # Skip the first row which contains the header description
    state_mapping_df = state_mapping_df[state_mapping_df['value'] != 'Label:'].copy()
    state_mapping_df['value'] = pd.to_numeric(state_mapping_df['value'], errors='coerce')
    state_mapping_df = state_mapping_df.dropna(subset=['value'])
    state_mapping_df['value'] = state_mapping_df['value'].astype(int)
    
    # Create a dictionary for mapping FIPS code to state name
    state_fips_to_name = dict(zip(state_mapping_df['value'], state_mapping_df['label']))
    
    print(f"âœ“ Loaded state FIPS mapping")
    print(f"  Total states/territories: {len(state_fips_to_name)}")
    print(f"\nSample mappings:")
    for fips, name in list(state_fips_to_name.items())[:5]:
        print(f"  {fips:2d} â†’ {name}")
else:
    print(f"âœ— File not found: {state_mapping_path}")
    state_fips_to_name = {}
```

```python
# Apply state name mapping to CDC data
if 'cdc_df' in globals() and state_fips_to_name:
    # Create new 'Location' column in CDC data by mapping _STATE to state names
    cdc_df['Location'] = cdc_df['_STATE'].map(state_fips_to_name)
    
    print(f"âœ“ Added 'Location' column to CDC data")
    print(f"  Mapped {cdc_df['Location'].notna().sum():,} out of {len(cdc_df):,} rows")
    print(f"  Unmapped FIPS codes: {cdc_df['Location'].isna().sum()}")
    
    # Show sample of the mapping
    print(f"\nSample CDC data with Location:")
    display(cdc_df[['_STATE', 'Location', 'IYEAR', 'Age_Group']].head(10))
    
    # Check unique locations in both datasets
    if 'kff_df' in globals():
        kff_locations = set(kff_df['Location'].unique())
        cdc_locations = set(cdc_df['Location'].dropna().unique())
        common_locations = kff_locations & cdc_locations
        
        print(f"\nðŸ“Š Location overlap analysis:")
        print(f"  KFF unique locations: {len(kff_locations)}")
        print(f"  CDC unique locations: {len(cdc_locations)}")
        print(f"  Common locations: {len(common_locations)}")
        
        # Show locations that don't overlap
        kff_only = kff_locations - cdc_locations
        cdc_only = cdc_locations - kff_locations
        if kff_only:
            print(f"\n  KFF-only locations: {sorted(kff_only)}")
        if cdc_only:
            print(f"  CDC-only locations: {sorted(cdc_only)}")
else:
    print("âš  CDC data or state mapping not available")
```

```python
# Check common dimensions for potential joins
if 'kff_df' in globals() and 'cdc_df' in globals():
    print("=" * 100)
    print("Data Compatibility Analysis")
    print("=" * 100)
    
    # Check for common columns
    kff_cols = set(kff_df.columns)
    brfss_cols = set(cdc_df.columns)
    common_cols = kff_cols & brfss_cols
    
    print(f"\nCommon columns: {common_cols if common_cols else 'None'}")
    
    # Check for potential join keys
    potential_keys = ['Year', 'year', 'State', 'state', 'Location', 'location', '_STATE', 'StateCode']
    print(f"\nPotential join keys:")
    print(f"  KFF columns: {[col for col in kff_df.columns if any(k.lower() in col.lower() for k in potential_keys)]}")
    print(f"  CDC columns: {[col for col in cdc_df.columns if any(k.lower() in col.lower() for k in potential_keys)]}")
    
    # Check year ranges
    kff_year_col = [col for col in kff_df.columns if 'year' in col.lower()]
    brfss_year_col = [col for col in cdc_df.columns if 'year' in col.lower()]
    
    if kff_year_col and brfss_year_col:
        kff_years = sorted(kff_df[kff_year_col[0]].dropna().unique())
        brfss_years = sorted(cdc_df[brfss_year_col[0]].dropna().unique())
        overlapping_years = set(kff_years) & set(brfss_years)
        
        print(f"\nYear ranges:")
        print(f"  KFF: {kff_years}")
        print(f"  CDC: {brfss_years}")
        print(f"  Overlapping years: {sorted(overlapping_years) if overlapping_years else 'None'}")
```

# Part 2: Time-Lagged Prediction Model

## Objective
Build a predictive model where **CDC health risk data from Year T** predicts **KFF premium prices in Year T+2**.

### Business Logic
Insurance companies use historical health data to forecast future risk and set premiums with a ~2-year lag.

## 7. Create Time-Lagged Dataset (T â†’ T+2)

```python
# Step 1: Prepare CDC data (X features) with Year column for T â†’ T+2 mapping
if 'cdc_df' in globals():
    # Ensure we have the renamed Year column
    cdc_for_modeling = cdc_df.copy()
    if 'IYEAR' in cdc_for_modeling.columns and 'Year' not in cdc_for_modeling.columns:
        cdc_for_modeling['Year'] = cdc_for_modeling['IYEAR']
    
    # Select health risk features (X)
    health_features = [
        'Location', 'Year', 'Age_Group', '_STATE',
        'mean_BMI_w', 
        'ever_smoker_prev_w', 
        'current_smoker_prev_w',
        'diabetes_prev_w', 
        'heart_attack_prev_w',
        'chd_prev_w', 
        'stroke_prev_w',
        'asthma_prev_w',
        'copd_prev_w',
        'arthritis_prev_w',
        'has_children_prev_w',
        'mean_num_adults_w',
        'mean_household_size_w'
    ]
    
    # Keep only available columns
    available_features = [col for col in health_features if col in cdc_for_modeling.columns]
    cdc_features = cdc_for_modeling[available_features].copy()
    
    print(f"âœ“ Prepared CDC features for modeling")
    print(f"  Shape: {cdc_features.shape}")
    print(f"  Year range: {cdc_features['Year'].min()} - {cdc_features['Year'].max()}")
    print(f"  Available features: {len(available_features)}")
    display(cdc_features.head())
else:
    print("âš  CDC data not loaded")
```

```python
# Step 2: Prepare KFF data (Y target) - use Average Benchmark Premium as target
if 'kff_df' in globals():
    kff_target = kff_df.copy()
    
    # Clean KFF premium columns - remove '$' and convert to float
    premium_cols = [
        'Average Lowest-Cost Bronze Premium',
        'Average Lowest-Cost Silver Premium', 
        'Average Benchmark Premium',
        'Average Lowest-Cost Gold Premium'
    ]
    
    for col in premium_cols:
        if col in kff_target.columns:
            # Remove $ and convert to numeric
            kff_target[col] = kff_target[col].str.replace('$', '', regex=False).str.replace(',', '')
            kff_target[col] = pd.to_numeric(kff_target[col], errors='coerce')
    
    # Use Average Benchmark Premium as our primary target (Y)
    kff_target = kff_target[['Location', 'Year', 'Average Benchmark Premium']].copy()
    kff_target = kff_target.rename(columns={'Average Benchmark Premium': 'Premium_Y'})
    
    print(f"âœ“ Prepared KFF target variable (Y)")
    print(f"  Shape: {kff_target.shape}")
    print(f"  Year range: {kff_target['Year'].min()} - {kff_target['Year'].max()}")
    print(f"\nPremium statistics:")
    print(kff_target['Premium_Y'].describe())
    display(kff_target.head())
else:
    print("âš  KFF data not loaded")
```

```python
# Step 3: Create time-lagged join (CDC Year T â†’ KFF Year T+2)
if 'cdc_features' in globals() and 'kff_target' in globals():
    # Create a copy of CDC data and add 2 years for the lag
    cdc_lagged = cdc_features.copy()
    cdc_lagged['Year_Target'] = cdc_lagged['Year'] + 2  # T â†’ T+2
    
    # Rename columns before merge to avoid conflicts
    cdc_lagged = cdc_lagged.rename(columns={'Year': 'CDC_Year'})
    kff_target_renamed = kff_target.rename(columns={'Year': 'KFF_Year'})
    
    # Join CDC(T) with KFF(T+2)
    modeling_df = pd.merge(
        cdc_lagged,
        kff_target_renamed,
        left_on=['Location', 'Year_Target'],
        right_on=['Location', 'KFF_Year'],
        how='inner'
    )
    
    # Drop redundant column
    modeling_df = modeling_df.drop(columns=['Year_Target'])
    
    print(f"âœ“ Created time-lagged dataset (CDC Year T â†’ KFF Year T+2)")
    print(f"  Total rows: {len(modeling_df):,}")
    print(f"  Total columns: {modeling_df.shape[1]}")
    
    # Show the time alignment
    print(f"\nðŸ“Š Time Alignment Analysis:")
    alignment_summary = modeling_df.groupby(['CDC_Year', 'KFF_Year']).size().reset_index(name='count')
    print(alignment_summary.to_string(index=False))
    
    print(f"\nâœ“ Sample of time-lagged data:")
    display(modeling_df[['Location', 'CDC_Year', 'KFF_Year', 'Age_Group', 
                         'mean_BMI_w', 'diabetes_prev_w', 'Premium_Y']].head(10))
    
    # Save to global
    globals()['modeling_df'] = modeling_df
    
else:
    print("âš  CDC features or KFF target not prepared")
```

## 8. Split into Train/Validation/Test Sets

```python
# Split data by KFF year (the target year)
# Train: KFF 2022, 2023, 2024 (from CDC 2020, 2021, 2022)
# Validation: KFF 2025 (from CDC 2023)
# Test: KFF 2026 (from CDC 2024)

if 'modeling_df' in globals():
    # Training set: KFF years 2022, 2023, 2024
    train_df = modeling_df[modeling_df['KFF_Year'].isin([2022, 2023, 2024])].copy()
    
    # Validation set: KFF year 2025
    val_df = modeling_df[modeling_df['KFF_Year'] == 2025].copy()
    
    # Test set: KFF year 2026
    test_df = modeling_df[modeling_df['KFF_Year'] == 2026].copy()
    
    print("=" * 100)
    print("ðŸ“Š Dataset Split Summary")
    print("=" * 100)
    
    print(f"\nðŸ”¹ Training Set:")
    print(f"   KFF Years: 2022, 2023, 2024 (from CDC 2020, 2021, 2022)")
    print(f"   Rows: {len(train_df):,}")
    print(f"   CDC Years: {sorted(train_df['CDC_Year'].unique())}")
    print(f"   States: {len(train_df['Location'].unique())}")
    
    print(f"\nðŸ”¹ Validation Set:")
    print(f"   KFF Year: 2025 (from CDC 2023)")
    print(f"   Rows: {len(val_df):,}")
    print(f"   CDC Years: {sorted(val_df['CDC_Year'].unique())}")
    print(f"   States: {len(val_df['Location'].unique())}")
    
    print(f"\nðŸ”¹ Test Set (Critical for Rate Adequacy):")
    print(f"   KFF Year: 2026 (from CDC 2024)")
    print(f"   Rows: {len(test_df):,}")
    print(f"   CDC Years: {sorted(test_df['CDC_Year'].unique())}")
    print(f"   States: {len(test_df['Location'].unique())}")
    
    print(f"\nâœ“ Total samples: {len(train_df) + len(val_df) + len(test_df):,}")
    
    # Save to globals
    globals()['train_df'] = train_df
    globals()['val_df'] = val_df
    globals()['test_df'] = test_df
    
else:
    print("âš  Modeling dataframe not available")
```

## 9. Feature Engineering & Preprocessing

```python
# Define feature columns for the model
if 'train_df' in globals():
    # Identify numerical health features
    numerical_features = [
        'mean_BMI_w',
        'ever_smoker_prev_w',
        'current_smoker_prev_w', 
        'diabetes_prev_w',
        'heart_attack_prev_w',
        'chd_prev_w',
        'stroke_prev_w',
        'asthma_prev_w',
        'copd_prev_w',
        'arthritis_prev_w',
        'has_children_prev_w',
        'mean_num_adults_w',
        'mean_household_size_w'
    ]
    
    # Keep only features that exist in the data
    available_numerical = [f for f in numerical_features if f in train_df.columns]
    
    # Categorical features
    categorical_features = ['Age_Group']
    
    # All features (X)
    feature_cols = available_numerical + categorical_features
    
    # Target variable (Y)
    target_col = 'Premium_Y'
    
    print("=" * 100)
    print("ðŸ”§ Feature Engineering Configuration")
    print("=" * 100)
    
    print(f"\nðŸ“‹ Selected Features ({len(feature_cols)} total):")
    print(f"\n  Numerical features ({len(available_numerical)}):")
    for i, feat in enumerate(available_numerical, 1):
        print(f"    {i:2d}. {feat}")
    
    print(f"\n  Categorical features ({len(categorical_features)}):")
    for i, feat in enumerate(categorical_features, 1):
        print(f"    {i:2d}. {feat}")
    
    print(f"\nðŸŽ¯ Target variable: {target_col}")
    
    # Check for missing values
    print(f"\nðŸ“Š Missing Value Analysis (Training Set):")
    missing_summary = train_df[feature_cols + [target_col]].isnull().sum()
    missing_pct = (missing_summary / len(train_df) * 100).round(2)
    missing_df = pd.DataFrame({
        'Feature': missing_summary.index,
        'Missing Count': missing_summary.values,
        'Missing %': missing_pct.values
    })
    missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing Count', ascending=False)
    
    if len(missing_df) > 0:
        print(missing_df.to_string(index=False))
    else:
        print("  âœ“ No missing values found!")
    
else:
    print("âš  Training data not available")
```

```python
# Preprocessing: Handle missing values and prepare features
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

if 'train_df' in globals() and 'feature_cols' in globals():
    # Create preprocessing pipeline
    print("=" * 100)
    print("ðŸ”§ Preprocessing Pipeline")
    print("=" * 100)
    
    # Step 1: Handle missing values with median imputation for numerical features
    imputer = SimpleImputer(strategy='median')
    
    # Fit on training data
    X_train_num = train_df[available_numerical].copy()
    X_train_num_imputed = pd.DataFrame(
        imputer.fit_transform(X_train_num),
        columns=available_numerical
    )
    
    # Transform validation and test sets
    X_val_num = val_df[available_numerical].copy()
    X_val_num_imputed = pd.DataFrame(
        imputer.transform(X_val_num),
        columns=available_numerical
    )
    
    X_test_num = test_df[available_numerical].copy()
    X_test_num_imputed = pd.DataFrame(
        imputer.transform(X_test_num),
        columns=available_numerical
    )
    
    print(f"\nâœ“ Missing values imputed with median strategy")
    
    # Step 2: Standardize numerical features
    scaler = StandardScaler()
    
    X_train_scaled = pd.DataFrame(
        scaler.fit_transform(X_train_num_imputed),
        columns=available_numerical
    )
    
    X_val_scaled = pd.DataFrame(
        scaler.transform(X_val_num_imputed),
        columns=available_numerical
    )
    
    X_test_scaled = pd.DataFrame(
        scaler.transform(X_test_num_imputed),
        columns=available_numerical
    )
    
    print(f"âœ“ Features standardized (mean=0, std=1)")
    
    # Step 3: Add categorical features (Age_Group) - keep indices aligned
    X_train_final = pd.concat([X_train_scaled, train_df[['Age_Group']].reset_index(drop=True)], axis=1)
    X_val_final = pd.concat([X_val_scaled, val_df[['Age_Group']].reset_index(drop=True)], axis=1)
    X_test_final = pd.concat([X_test_scaled, test_df[['Age_Group']].reset_index(drop=True)], axis=1)
    
    # Extract target variables
    y_train = train_df[target_col].values
    y_val = val_df[target_col].values
    y_test = test_df[target_col].values
    
    print(f"\nðŸ“Š Preprocessed Data Summary:")
    print(f"  X_train: {X_train_final.shape}")
    print(f"  X_val:   {X_val_final.shape}")
    print(f"  X_test:  {X_test_final.shape}")
    print(f"  y_train: {y_train.shape}")
    print(f"  y_val:   {y_val.shape}")
    print(f"  y_test:  {y_test.shape}")
    
    print(f"\nâœ“ Data ready for model training!")
    
    # Save to globals
    globals()['X_train'] = X_train_final
    globals()['X_val'] = X_val_final
    globals()['X_test'] = X_test_final
    globals()['y_train'] = y_train
    globals()['y_val'] = y_val
    globals()['y_test'] = y_test
    globals()['scaler'] = scaler
    globals()['imputer'] = imputer
    
else:
    print("âš  Training data or feature columns not available")
```

## 10. Model Training - Random Forest Regressor

```python
# Train Random Forest model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import time

if 'X_train' in globals() and 'y_train' in globals():
    print("=" * 100)
    print("ðŸ¤– Training Random Forest Regressor")
    print("=" * 100)
    
    # Initialize model
    rf_model = RandomForestRegressor(
        n_estimators=100,
        max_depth=15,
        min_samples_split=5,
        min_samples_leaf=2,
        random_state=42,
        n_jobs=-1,
        verbose=0
    )
    
    print(f"\nðŸ“‹ Model Configuration:")
    print(f"  n_estimators: 100")
    print(f"  max_depth: 15")
    print(f"  min_samples_split: 5")
    print(f"  min_samples_leaf: 2")
    
    # Train model
    print(f"\nðŸ”„ Training model...")
    start_time = time.time()
    rf_model.fit(X_train, y_train)
    training_time = time.time() - start_time
    
    print(f"âœ“ Training completed in {training_time:.2f} seconds")
    
    # Make predictions on all sets
    y_train_pred = rf_model.predict(X_train)
    y_val_pred = rf_model.predict(X_val)
    y_test_pred = rf_model.predict(X_test)
    
    # Calculate metrics
    def calculate_metrics(y_true, y_pred, set_name):
        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        r2 = r2_score(y_true, y_pred)
        mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
        
        print(f"\nðŸ“Š {set_name} Performance:")
        print(f"  MAE:  ${mae:,.2f}")
        print(f"  RMSE: ${rmse:,.2f}")
        print(f"  RÂ²:   {r2:.4f}")
        print(f"  MAPE: {mape:.2f}%")
        
        return {'MAE': mae, 'RMSE': rmse, 'R2': r2, 'MAPE': mape}
    
    print("\n" + "=" * 100)
    train_metrics = calculate_metrics(y_train, y_train_pred, "Training Set")
    val_metrics = calculate_metrics(y_val, y_val_pred, "Validation Set")
    test_metrics = calculate_metrics(y_test, y_test_pred, "Test Set (2026)")
    
    # Save to globals
    globals()['rf_model'] = rf_model
    globals()['y_train_pred'] = y_train_pred
    globals()['y_val_pred'] = y_val_pred
    globals()['y_test_pred'] = y_test_pred
    
    print(f"\nâœ“ Model and predictions saved!")
    
else:
    print("âš  Preprocessed data not available")
```

## 11. Rate Adequacy Analysis (2026 Pricing Risk Assessment)

```python
# Analyze pricing adequacy: Compare predicted vs actual 2026 premiums
if 'y_test_pred' in globals() and 'test_df' in globals():
    # Create analysis dataframe
    adequacy_df = test_df[['Location', 'CDC_Year', 'KFF_Year', 'Age_Group', '_STATE']].copy()
    adequacy_df['Actual_Premium'] = y_test
    adequacy_df['Predicted_Premium'] = y_test_pred
    adequacy_df['Premium_Diff'] = adequacy_df['Predicted_Premium'] - adequacy_df['Actual_Premium']
    adequacy_df['Diff_Pct'] = (adequacy_df['Premium_Diff'] / adequacy_df['Actual_Premium']) * 100
    
    # Classify pricing risk
    def classify_risk(diff_pct):
        if diff_pct > 10:
            return 'High Underprice Risk (Loss)'
        elif diff_pct > 5:
            return 'Moderate Underprice Risk'
        elif diff_pct < -10:
            return 'High Overprice Risk (Competition)'
        elif diff_pct < -5:
            return 'Moderate Overprice Risk'
        else:
            return 'Adequate Pricing'
    
    adequacy_df['Risk_Category'] = adequacy_df['Diff_Pct'].apply(classify_risk)
    
    print("=" * 100)
    print("ðŸ’° Rate Adequacy Analysis - 2026 Pricing Risk Assessment")
    print("=" * 100)
    
    print(f"\nðŸ“Š Overall Statistics:")
    print(f"  Total observations: {len(adequacy_df):,}")
    print(f"  Mean actual premium: ${adequacy_df['Actual_Premium'].mean():,.2f}")
    print(f"  Mean predicted premium: ${adequacy_df['Predicted_Premium'].mean():,.2f}")
    print(f"  Average difference: ${adequacy_df['Premium_Diff'].mean():,.2f}")
    print(f"  Average % difference: {adequacy_df['Diff_Pct'].mean():.2f}%")
    
    # Risk distribution
    print(f"\nâš ï¸  Pricing Risk Distribution:")
    risk_counts = adequacy_df['Risk_Category'].value_counts()
    for risk, count in risk_counts.items():
        pct = (count / len(adequacy_df)) * 100
        print(f"  {risk}: {count:,} ({pct:.1f}%)")
    
    # Top states with underpricing risk (loss risk for insurers)
    print(f"\nðŸ”´ Top 10 States with UNDERPRICING Risk (Predicted > Actual):")
    state_risk = adequacy_df.groupby('Location').agg({
        'Premium_Diff': 'mean',
        'Diff_Pct': 'mean',
        'Actual_Premium': 'mean',
        'Predicted_Premium': 'mean'
    }).round(2)
    
    underpriced = state_risk[state_risk['Premium_Diff'] > 0].sort_values('Premium_Diff', ascending=False).head(10)
    print(underpriced.to_string())
    
    # Top states with overpricing risk (competition risk)
    print(f"\nðŸŸ¢ Top 10 States with OVERPRICING Risk (Predicted < Actual):")
    overpriced = state_risk[state_risk['Premium_Diff'] < 0].sort_values('Premium_Diff').head(10)
    print(overpriced.to_string())
    
    # Save to globals
    globals()['adequacy_df'] = adequacy_df
    
else:
    print("âš  Test predictions not available")
```
